# Adicionar, Remover e Modificar Itens no Dicionário

### Estrutura:

- Adicionar itens:

```python
lucro_1tri = {'janeiro': 100000, 'fevereiro': 120000, 'março': 90000}
lucro_2tri = {'abril': 88000, 'maio': 89000, 'junho': 120000}
#adicionando 1 item

#adicionando vários itens ou um dicionário a outro

#adicionando um item já existente (manualmente ou pelo update)
```

```python
# PLANEJAMENTO DA ATUALIZAÇÃO DE LUCROS:
# 1. Definir os dicionários bases 'lucro_1tri' e 'lucro_2tri'.
# 2. CENÁRIO 1 (Adicionar 1 item manualmente):
#    - Usar a sintaxe de colchetes dict[chave] = valor para incluir o mês de 'julho' no 'lucro_2tri'.
# 3. CENÁRIO 2 (Adicionar vários itens ou um dicionário a outro):
#    - Usar o método .update() para juntar todos os meses do 'lucro_2tri' dentro do 'lucro_1tri'.
# 4. CENÁRIO 3 (Adicionar/Modificar um item já existente):
#    - Mostrar que tentar atribuir um novo valor para 'janeiro' (manualmente ou por update) altera o dado existente em vez de criar uma duplicata.
# 5. Exibir os resultados de cada etapa na tela.

# 1. Dicionários base fornecidos
lucro_1tri = {'janeiro': 100000, 'fevereiro': 120000, 'março': 90000}
lucro_2tri = {'abril': 88000, 'maio': 89000, 'junho': 120000}

print(f"{' MANIPULAÇÃO DE DICIONÁRIOS ':=^50}\n")

# ------------------------------------------------------------------------------
# Cenário 1: Adicionando 1 item (Sintaxe Direta)
# ------------------------------------------------------------------------------
# Se a chave 'julho' não existe no lucro_2tri, o Python a cria na hora
lucro_2tri['julho'] = 95000

print("1. Item 'julho' adicionado ao 2º Trimestre:")
print(lucro_2tri)
print()


# ------------------------------------------------------------------------------
# Cenário 2: Adicionando vários itens ou um dicionário a outro (.update)
# ------------------------------------------------------------------------------
# Vamos "jogar" tudo o que está dentro de lucro_2tri para dentro do lucro_1tri
lucro_1tri.update(lucro_2tri)

print("2. 1º Trimestre após absorver o 2º Trimestre com .update():")
print(lucro_1tri)
print()


# ------------------------------------------------------------------------------
# Cenário 3: Adicionando um item já existente (Manualmente e por .update)
# ------------------------------------------------------------------------------
# Tentando atualizar 'janeiro' manualmente (de 100000 para 110000)
lucro_1tri['janeiro'] = 110000

# Tentando atualizar 'fevereiro' usando o .update() em lote
lucro_1tri.update({'fevereiro': 125000, 'agosto': 102000})

print("3. Dicionário final após modificações de chaves existentes:")
print(lucro_1tri)

print(f"\n{'-' * 50}")
```

- Modificar itens:

Da mesma forma que adicionamos 1 valor, caso essa chave já exista o item é apenas modificado.

dicionario[chave] = valor

Vamos modificar o lucro de fevereiro:<br>
(Lembrando: caso o item não exista, ele vai criar o item no dicionário)

```python
lucro_fev = 85000
```

```python
# PLANEJAMENTO DA ATUALIZAÇÃO MANUAL:
# 1. Definir o dicionário base 'lucro_1tri' com os dados iniciais.
# 2. Definir a variável 'lucro_fev' com o novo valor que queremos aplicar (85000).
# 3. Aplicar a modificação direta usando a sintaxe dict[chave] = valor.
# 4. Exibir o dicionário antes e depois da alteração para comparar o resultado.

# 1. Dicionário inicial
lucro_1tri = {'janeiro': 100000, 'fevereiro': 120000, 'março': 90000}

print(f"{' ANTES DA MODIFICAÇÃO ':-^45}")
print(lucro_1tri)

# 2. Novo valor definido
lucro_fev = 85000

# 3. Modificando o valor da chave 'fevereiro'
# O Python vai até a chave 'fevereiro' e troca o 120000 por 85000
lucro_1tri['fevereiro'] = lucro_fev

print(f"\n{' DEPOIS DA MODIFICAÇÃO ':-^45}")
print(lucro_1tri)
print("-" * 45)
```

- Remover itens:

del dicionario[chave]

ou então

valor = dicionario.pop(chave)

mas cuidado com:

```python
#removendo o mês de junho


#obs: o del também funciona para listas, caso queira usar
#del lista[i]
funcionarios = ['João', 'Lira', 'Maria', 'Ana', 'Paula']
```

```python
# PLANEJAMENTO DA REMOÇÃO:
# 1. Definir o dicionário base de lucros com o mês de 'junho'.
# 2. Definir a lista de funcionários para testar o 'del' com índices.
# 3. CENÁRIO 1 (Usando .pop()):
#    - Remover 'junho' do dicionário isolando o valor removido em uma variável.
# 4. CENÁRIO 2 (Usando del no dicionário):
#    - Remover o mês de 'maio' usando a sintaxe nativa 'del'.
# 5. CENÁRIO 3 (Usando del na lista):
#    - Remover o funcionário 'Maria' usando o 'del' com o índice correspondente (índice 2).
# 6. Exibir os resultados de cada etapa.

# 1. Bases de dados para o teste
lucro_2tri = {'abril': 88000, 'maio': 89000, 'junho': 120000}
funcionarios = ['João', 'Lira', 'Maria', 'Ana', 'Paula']

print(f"{' ESTRUTURAS ORIGINAIS ':-^45}")
print(f"Dicionário: {lucro_2tri}")
print(f"Lista:       {funcionarios}\n")

# ------------------------------------------------------------------------------
# Cenário 1: Removendo com .pop(chave)
# ------------------------------------------------------------------------------
# O .pop() remove o item e te devolve o valor, caso você precise usar para algo
lucro_junho_removido = lucro_2tri.pop('junho')

print(f"{' REMOÇÃO COM .POP() ':-^45}")
print(f"Valor que foi retirado de junho: R$ {lucro_junho_removido}")
print(f"Dicionário após o .pop(): {lucro_2tri}\n")


# ------------------------------------------------------------------------------
# Cenário 2: Removendo com del dicionario[chave]
# ------------------------------------------------------------------------------
# O 'del' apenas deleta o par direto na memória, sem retornar valor nenhum
del lucro_2tri['maio']

print(f"{' REMOÇÃO COM DEL NO DICT ':-^45}")
print(f"Dicionário após o 'del': {lucro_2tri}\n")


# ------------------------------------------------------------------------------
# Cenário 3: Usando o del para listas (del lista[i])
# ------------------------------------------------------------------------------
# 'Maria' está no índice 2 da nossa lista (João=0, Lira=1, Maria=2)
del funcionarios[2]

print(f"{' REMOÇÃO COM DEL NA LISTA ':-^45}")
print(f"Lista após deletar o índice 2 (Maria): {funcionarios}")
print("-" * 45)
```
