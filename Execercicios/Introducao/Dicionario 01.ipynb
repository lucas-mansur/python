# Dicionários em Python

### Estrutura:

dicionario = {chave: valor, chave: valor, chave: valor, chave: valor ...}

### Vantagens e Desvantagens

- Não devem ser usados para pegar itens em uma determinada ordem
- Podem ter valores heterogêneos (vários tipos de valores dentro de um mesmo dicionário: inteiros, strings, listas, etc)
- Chaves são únicas obrigatoriamente
- Mais intuitivos de trabalhar

```python
mais_vendidos = {'tecnologia': 'iphone', 'refrigeracao': 'ar consul 12000 btu', 'livros': 'o alquimista', 'eletrodoméstico': 'geladeira', 'lazer': 'prancha surf'}

vendas_tecnologia = {'iphone': 15000, 'samsung galaxy': 12000, 'tv samsung': 10000, 'ps5': 14300, 'tablet': 1720, 'ipad': 1000, 'tv philco': 2500, 'notebook hp': 1000, 'notebook dell': 17000, 'notebook asus': 2450}
```

- Qual foi o item mais vendido nas categorias 'livros' e 'lazer'?
- Quanto foi vendido de 'notebook asus' e de 'ipad'?

```python
# PLANEJAMENTO DA ANÁLISE DE PRODUTOS E VENDAS:
# 1. Definir os dois dicionários com os dados fornecidos ('mais_vendidos' e 'vendas_tecnologia').
# 2. Pergunta 1: Buscar diretamente no dicionário 'mais_vendidos' usando as chaves 'livros' e 'lazer'.
# 3. Pergunta 2: Buscar a quantidade de vendas no dicionário 'vendas_tecnologia' usando as chaves 'notebook asus' e 'ipad'.
# 4. Armazenar os resultados em variáveis com nomes claros.
# 5. Exibir os resultados na tela de forma organizada e pythônica.


# 1. Base de dados do e-commerce
mais_vendidos = {
    'tecnologia': 'iphone', 
    'refrigeracao': 'ar consul 12000 btu', 
    'livros': 'o alquimista', 
    'eletrodoméstico': 'geladeira', 
    'lazer': 'prancha surf'
}

vendas_tecnologia = {
    'iphone': 15000, 
    'samsung galaxy': 12000, 
    'tv samsung': 10000, 
    'ps5': 14300, 
    'tablet': 1720, 
    'ipad': 1000, 
    'tv philco': 2500, 
    'notebook hp': 1000, 
    'notebook dell': 17000, 
    'notebook asus': 2450
}

# 2. Respondendo à primeira pergunta (Acesso direto por Chave)
livro_mais_vendido = mais_vendidos['livros']
lazer_mais_vendido = mais_vendidos['lazer']

# 3. Respondendo à segunda pergunta (Buscando os valores de vendas)
vendas_asus = vendas_tecnologia['notebook asus']
vendas_ipad = vendas_tecnologia['ipad']

# 5. Exibição dos Relatórios Formatados
print(f"{' ANÁLISE DE MAIS VENDIDOS ':=^45}\n")
print(f"O item mais vendido em 'livros' foi: {livro_mais_vendido.title()}")
print(f"O item mais vendido em 'lazer' foi:  {lazer_mais_vendido.title()}")

print(f"\n{' DETALHE DE VENDAS (TECNOLOGIA) ':=^45}\n")
print(f"Quantidade vendida de Notebook Asus: {vendas_asus} unidades")
print(f"Quantidade vendida de iPad:          {vendas_ipad} unidades")
print(f"\n{'-' * 45}")
```
