# Aplicação de Tupla - Lista de Tuplas

### Estrutura:

Além de casos como o do enumerate, em que usamos uma função para transformar itens em tuplas porque isso ajuda o nosso código, temos também listas de tuplas como algo comum dentro do Python.

### Exemplo:

Digamos que você está analisando as vendas do Banco de Dados de um e-commerce.

Em um determinado dia, você extraiu as vendas do Banco de Dados e elas vieram nesse formato:

```python
vendas = [
    ('20/08/2020', 'iphone x', 'azul', '128gb', 350, 4000),
    ('20/08/2020', 'iphone x', 'prata', '128gb', 1500, 4000),
    ('20/08/2020', 'ipad', 'prata', '256gb', 127, 6000),
    ('20/08/2020', 'ipad', 'prata', '128gb', 981, 5000),
    ('21/08/2020', 'iphone x', 'azul', '128gb', 397, 4000),
    ('21/08/2020', 'iphone x', 'prata', '128gb', 1017, 4000),
    ('21/08/2020', 'ipad', 'prata', '256gb', 50, 6000),
    ('21/08/2020', 'ipad', 'prata', '128gb', 4000, 5000),
]
```

- Qual foi o faturamento de IPhone no dia 20/08/2020?
- Qual foi o produto mais vendido (em unidades) no dia 21/08/2020?

```python
# PLANEJAMENTO DA ANÁLISE DE FATURAMENTO:
# 1. Criar uma variável acumuladora 'faturamento_iphone_20' e inicializá-la com 0.0.
# 2. Criar um loop 'for' para percorrer cada registro da lista de vendas.
# 3. Usar o DESEMPACOTAMENTO DE TUPLAS para dar nomes claros às colunas (data, produto, cor, capacidade, unidades, preco).
# 4. Criar uma estrutura de decisão 'if' com duas condições:
#    - A data deve ser exatamente igual a '20/08/2020'.
#    - O nome do produto deve conter a palavra 'iphone' (usando o operador 'in').
# 5. Se a linha atender aos critérios, calcular o faturamento daquela venda: (unidades * preco).
# 6. Somar esse valor calculado ao acumulador 'faturamento_iphone_20'.
# 7. Fora do loop, exibir o faturamento total formatado com estilo pythônico.

vendas = [
    ('20/08/2020', 'iphone x', 'azul', '128gb', 350, 4000),
    ('20/08/2020', 'iphone x', 'prata', '128gb', 1500, 4000),
    ('20/08/2020', 'ipad', 'prata', '256gb', 127, 6000),
    ('20/08/2020', 'ipad', 'prata', '128gb', 981, 5000),
    ('21/08/2020', 'iphone x', 'azul', '128gb', 397, 4000),
    ('21/08/2020', 'iphone x', 'prata', '128gb', 1017, 4000),
    ('21/08/2020', 'ipad', 'prata', '256gb', 50, 6000),
    ('21/08/2020', 'ipad', 'prata', '128gb', 4000, 5000),
]

faturamento_iphone_20 = 0.0

# Executando o loop com desempacotamento conforme planejado
for data, produto, cor, capacidade, unidades, preco in vendas:
    # Aplicando os filtros do passo 5
    if data == '20/08/2020' and 'iphone' in produto:
        # Passos 6 e 7: Cálculo e acumulação
        faturamento_linha = unidades * preco
        faturamento_iphone_20 += faturamento_linha

print(f"{' Relatório de Faturamento ':-^45}")
print(f"Faturamento de iPhone em 20/08/2020: R$ {faturamento_iphone_20:,.2f}")
print(f"{'-' * 45}")
```

```python
# PLANEJAMENTO DA ANÁLISE DE MAIOR VOLUME:
# 1. Criar uma variável 'produto_mais_vendido' inicializada como uma string vazia ("").
# 2. Criar uma variável 'maior_quantidade' inicializada com 0 para guardar o recorde de unidades.
# 3. Criar o loop 'for' para percorrer a lista de vendas.
# 4. Desempacotar a tupla da rodada em variáveis legíveis.
# 5. Criar um 'if' para filtrar apenas os registros onde a data seja exatamente '21/08/2020'.
# 6. Dentro desse filtro de data, criar outro 'if' para validar o recorde:
#    - Verificar se as 'unidades' da linha atual são MAIORES do que o valor guardado em 'maior_quantidade'.
# 7. Se for maior, atualizar as duas variáveis de controle:
#    - 'maior_quantidade' recebe o novo número de unidades.
#    - 'produto_mais_vendido' recebe o nome do produto atual.
# 8. Fora do loop, exibir o nome do produto campeão e a quantidade de unidades dele.

produto_mais_vendido = ""
maior_quantidade = 0

# Executando o loop conforme o planejamento
for data, produto, cor, capacidade, unidades, preco in vendas:
    # Passo 5: Filtro de data
    if data == '21/08/2020':
        # Passo 6: Validação do recorde (algoritmo do maior)
        if unidades > maior_quantidade:
            # Passo 7: Atualização das variáveis campeãs
            maior_quantidade = unidades
            produto_mais_vendido = produto

print(f"\n{" Análise de Volume de Vendas ":-^45}")
print(f"O produto mais vendido em 21/08/2020 foi: {produto_mais_vendido.upper()}")
print(f"Volume vendido: {maior_quantidade} unidades")
print("-" * 45)
```
