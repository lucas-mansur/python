## 1. Exercício "menos prático" para treinar manipulação de dicionário

Dessa vez, vamos apenas treinar a manipulação de dicionário. Transforme as listas abaixo em 1 único dicionário no formato:

- Apesar de parecer "menos prático" esse é um procedimento que precisamos nos acostumar a fazer, visto que algumas funções (tema dos próximos módulos) precisam de dicionários para funcionar e saber transformar listas em dicionários (e vice-versa) é uma habilidade muito útil

Obs: Lembre do zip para juntar listas.<br>
Obs2: Repare que cada item das vendas é na verdade uma lista. Então é provável que você precise fazer esse código em 2 etapas

```python
produtos = ['iphone', 'galaxy', 'ipad', 'tv', 'máquina de café', 'kindle', 'geladeira', 'adega', 'notebook dell', 'notebook hp', 'notebook asus', 'microsoft surface', 'webcam', 'caixa de som', 'microfone', 'câmera canon']
vendas2019 = [558147,712350,573823,405252,718654,531580,973139,892292,422760,154753,887061,438508,237467,489705,328311,591120]
vendas2020 = [951642,244295,26964,787604,867660,78830,710331,646016,694913,539704,324831,667179,295633,725316,644622,994303]

#seu código aqui
```

```python
# PLANEJAMENTO DA CONSOLIDAÇÃO DE DADOS:
# 1. Analisar as três listas paralelas fornecidas (produtos, vendas2019 e vendas2020).
# 2. Criar um dicionário vazio chamado 'dicionario_consolidado = {}'.
# 3. Utilizar a função zip() para unir as TRÊS listas simultaneamente dentro de um loop 'for'.
# 4. Aplicar o UNPACKING no cabeçalho do 'for' para isolar as variáveis: 'prod', 'v19' e 'v20'.
# 5. A cada rodada do loop, criar uma nova chave no dicionário usando o nome do produto ('prod').
# 6. Atribuir a essa chave uma lista contendo os dois anos de vendas: '[v19, v20]'.
# 7. Exibir o dicionário final formatado e pronto.

# Listas originais fornecidas
produtos = ['iphone', 'galaxy', 'ipad', 'tv', 'máquina de café', 'kindle', 'geladeira', 'adega', 'notebook dell', 'notebook hp', 'notebook asus', 'microsoft surface', 'webcam', 'caixa de som', 'microfone', 'câmera canon']
vendas2019 = [558147,712350,573823,405252,718654,531580,973139,892292,422760,154753,887061,438508,237467,489705,328311,591120]
vendas2020 = [951642,244295,26964,787604,867660,78830,710331,646016,694913,539704,324831,667179,295633,725316,644622,994303]

# 2. Criando o dicionário vazio que receberá a estrutura
dicionario_consolidado = {}

# 3, 4, 5 e 6. O zip une os elementos de mesmo índice das 3 listas paralelas
for prod, v19, v20 in zip(produtos, vendas2019, vendas2020):
    # Criamos a chave com o produto e salvamos os valores em formato de lista [v19, v20]
    dicionario_consolidado[prod] = [v19, v20]

# 7. Exibindo o resultado final de forma estruturada para conferência
print(f"{' DICIONÁRIO CONSOLIDADO FINAL ':=^60}\n")

# Varrendo o novo dicionário para ver se ficou no formato exigido
for produto, dados_vendas in dicionario_consolidado.items():
    print(f"'{produto}': {dados_vendas}")

print(f"\n{'=' * 60}")
```

🧠 Conhecimento Agregado para o seu Arsenal:
Muitos programadores iniciantes acham que o zip() só consegue juntar duas listas por vez. Mas a verdade é que ele aceita quantas listas você quiser, desde que elas sejam paralelas (tenham o mesmo tamanho).

Se você fizesse em duas etapas usando apenas dict(zip()), você cairia em um problema, porque dict(zip(produtos, vendas2019)) criaria apenas a relação do primeiro ano. Ao rodar o processo para o segundo ano, você acabaria sobrescrevendo os valores antigos.

Fazendo com o loop for e injetando a lista [v19, v20] direto na chave, você montou uma estrutura multidimensional idêntica à que grandes bibliotecas de dados (como o Pandas) usam por baixo dos panos para criar tabelas!
