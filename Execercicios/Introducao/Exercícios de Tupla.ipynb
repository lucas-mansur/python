São exercícios bem parecidos com os que fizemos com listas. Mas na tupla, podemos não só trabalhar com índices, mas fazer o "unpacking" das tuplas, o que pode facilitar nossos códigos.

## 1. Análise de Vendas

Nesse exercício vamos fazer uma "análise simples" de atingimento de Meta.

Temos uma lista com os vendedores e os valores de vendas e queremos identificar (printar) quais os vendedores que bateram a meta e qual foi o valor que eles venderam.

```python
meta = 10000
vendas = [
    ('João', 15000),
    ('Julia', 27000),
    ('Marcus', 9900),
    ('Maria', 3750),
    ('Ana', 10300),
    ('Alon', 7870),
]
```

```python
# PLANEJAMENTO DA ANÁLISE DE METAS:
# 1. Definir o valor da meta estipulada (10000).
# 2. Criar a lista 'vendas' contendo as tuplas com o formato (vendedor, valor_venda).
# 3. Criar um cabeçalho pythônico para abrir o relatório de forma organizada.
# 4. Iniciar um loop 'for' para percorrer a lista de vendas.
# 5. Aplicar o UNPACKING de tuplas no próprio 'for' para extrair 'vendedor' e 'valor'.
# 6. Criar uma estrutura de decisão 'if' para verificar se o 'valor' é MAIOR ou IGUAL à 'meta'.
# 7. Se o vendedor tiver batido a meta, exibir o seu nome e o valor vendido formatado em R$.

# 1. Definição das variáveis e base de dados
meta = 10000
vendas = [
    ('João', 15000),
    ('Julia', 27000),
    ('Marcus', 9900),
    ('Maria', 3750),
    ('Ana', 10300),
    ('Alon', 7870),
]

# 3. Exibição do cabeçalho
print(f"{' Vendedores que Bateram a Meta ':-^45}")

# 4 e 5. Loop com Unpacking automático de cada tupla
for vendedor, valor in vendas:
    
    # 6. Validação da meta
    if valor >= meta:
        # 7. Exibição dos resultados válidos
        print(f"Sua equipe conta com: {vendedor:<10} | Vendeu: R$ {valor:,.2f}")

print(f"{'-' * 45}")
```

## 2. Comparação com Ano Anterior

Digamos que você está analisando as vendas de produtos de um ecommerce e quer identificar quais produtos tiveram no ano de 2020 mais vendas do que no ano de 2019, para reportar isso para a diretoria.

Sua resposta pode ser um print de cada produto, qual foi a venda de 2019, a venda de 2020 e o % de crescimento de 2020 para 2019.

Lembrando, para calcular o % de crescimento de um produto de um ano para o outro, podemos fazer: (vendas_produto2020/vendas_produto2019 - 1)

A lógica da tupla é: (produto, vendas2019, vendas2020)

```python
vendas_produtos = [('iphone', 558147, 951642), ('galaxy', 712350, 244295), ('ipad', 573823, 26964), ('tv', 405252, 787604), ('máquina de café', 718654, 867660), ('kindle', 531580, 78830), ('geladeira', 973139, 710331), ('adega', 892292, 646016), ('notebook dell', 422760, 694913), ('notebook hp', 154753, 539704), ('notebook asus', 887061, 324831), ('microsoft surface', 438508, 667179), ('webcam', 237467, 295633), ('caixa de som', 489705, 725316), ('microfone', 328311, 644622), ('câmera canon', 591120, 994303)]

#seu código aqui
```

```python
# PLANEJAMENTO DA ANÁLISE COMPARATIVA:
# 1. Definir a base de dados contendo as tuplas no formato (produto, vendas2019, vendas2020).
# 2. Exibir um cabeçalho pythônico estruturado para o relatório da diretoria.
# 3. Iniciar um loop 'for' para percorrer a lista 'vendas_produtos'.
# 4. Aplicar o UNPACKING para extrair as 3 colunas em variáveis claras: 'produto', 'vendas_19' e 'vendas_20'.
# 5. Criar uma estrutura de decisão 'if' para verificar se as vendas de 2020 foram MAIORES que as de 2019.
# 6. Se o produto teve crescimento:
#    - Calcular o percentual de crescimento usando a fórmula fornecida: (vendas_20 / vendas_19 - 1).
#    - Multiplicar por 100 para exibir em formato de porcentagem.
#    - Printar os dados do produto, vendas de 2019, vendas de 2020 e o % de crescimento formatado.

# 1. Base de dados fornecida
vendas_produtos = [
    ('iphone', 558147, 951642), 
    ('galaxy', 712350, 244295), 
    ('ipad', 573823, 26964), 
    ('tv', 405252, 787604), 
    ('máquina de café', 718654, 867660), 
    ('kindle', 531580, 78830), 
    ('geladeira', 973139, 710331), 
    ('adega', 892292, 646016), 
    ('notebook dell', 422760, 694913), 
    ('notebook hp', 154753, 539704), 
    ('notebook asus', 887061, 324831), 
    ('microsoft surface', 438508, 667179), 
    ('webcam', 237467, 295633), 
    ('caixa de som', 489705, 725316), 
    ('microfone', 328311, 644622), 
    ('câmera canon', 591120, 994303)
]

# 2. Exibição do cabeçalho com alinhamento das colunas do relatório
print(f"{' PRODUTOS COM CRESCIMENTO (2019 VS 2020) ':=^65}\n")
print(f"{'Produto':<20} | {'Vendas 2019':<12} | {'Vendas 2020':<12} | {'Crescimento'}")
print("-" * 65)

# 3 e 4. Loop com Unpacking automático dos 3 elementos da tupla
for produto, vendas_19, vendas_20 in vendas_produtos:
    
    # 5. Filtro: identificar apenas produtos que venderam mais em 2020
    if vendas_20 > vendas_19:
        
        # 6. Cálculo do indicador de crescimento (multiplicado por 100 para virar %)
        crescimento = (vendas_20 / vendas_19 - 1) * 100
        
        # Exibição da linha do relatório formatada com alinhamentos perfeitos
        print(f"{produto.title():<20} | {vendas_19:<12,} | {vendas_20:<12,} | {crescimento:+.1f}%")

print(f"\n{'-' * 65}")
```
