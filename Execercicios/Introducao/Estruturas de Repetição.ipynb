#### 6. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o maior faturamento

````python
print("Informe o seu faturamento dos últimos 5 meses, começando pelo mais recente, para que eu posso te informar qual é o maior")
mes = 1
# Inicializamos com None para que o primeiro faturamento digitado se torne o maior,
# não importa se ele for positivo, zero ou negativo (prejuízo).
maior = None
mes_maior = 1
while mes <= 5:
    faturamento = float(input(f"Informe o faturamento de {mes} atrás"))
    
    # Se for o primeiro mês OU se o faturamento atual for maior que o registrado
    if maior is None or faturamento > maior:
        maior = faturamento
        mes_maior = mes

    mes += 1
# Ajuste gramatical para o plural/singular de meses na resposta
texto_mes = "mês" if mes_maior == 1 else "meses"
print(f"\nO seu maior faturamento foi de R$ {maior:.2f} há {mes_maior} {texto_mes} atrás.")
````

#### 7. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o faturamento total (soma) e o faturamento médio por mês (média).

````python
print("Informe o seu faturamento dos últimos 5 meses para calcular o total e a média.")

mes = 1
total = 0.0  # Inicializa o acumulador como float

while mes <= 5:
    faturamento = float(input(f"Informe o faturamento do {mes}º mês: R$ "))
    
    # Acumulando o faturamento
    total += faturamento
    mes += 1

# Calculando a média (dividido por 5 meses)
media = total / 5

print(f"\n--- Resultado ---")
print(f"Faturamento Total: R$ {total:.2f}")
print(f"Faturamento Médio Mensal: R$ {media:.2f}")
````

#### 8. Faça um programa que consiga categorizar a idade das equipes de uma empresa. Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da equipe varia entre 0 e 25 (jovem) ,26 e 60 (sênior) e maior que 60 (idosa); e então, dizer se a equipe é jovem, sênior ou idosa, conforme a média calculada.

````python
print("Informe a quantidade de pessoas da equipe para classificar a sua equipe quanto à idade:")

# Validação para garantir que a equipe tenha pelo menos 1 pessoa
qtd = int(input("Informe quantas pessoas tem na sua equipe: "))
while qtd <= 0:
    print("A equipe deve ter pelo menos 1 pessoa.")
    qtd = int(input("Informe quantas pessoas tem na sua equipe: "))

aux = 0
soma_idade = 0

while aux < qtd:
    idade = int(input(f"Informe a idade da {aux + 1}ª pessoa: "))
    
    # Validação da idade: não deixa avançar enquanto não for uma idade realista
    while idade < 0 or idade > 120:
        print("Por favor, insira uma idade válida entre 0 e 120 anos.")
        idade = int(input(f"Informe a idade da {aux + 1}ª pessoa: "))
        
    soma_idade += idade
    aux += 1

media = soma_idade / qtd

if media <= 25:
    classificacao = "jovem"
elif media <= 60:
    classificacao = "sênior"
else:
    classificacao = "idosa"

# Usando o cabeçalho pythônico
print(f"\n{' Resultado ':-^35}")
print(f"Média de idade da equipe: {media:.1f} anos")
print(f"A sua equipe é considerada: {classificacao.upper()}")
````

#### 9. Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

````python
# variavel total_eleitores, tem que ser mair que 0
# variavel candidato_a,candidato_b e candidato_c
# loop no total_eleitores 
# input qual candidato 'a' 'b' ou 'c'
# somar ao 1 ao candidato escolhido
# informar a quantidade de votos para candidato

candidato_a = 0
candidato_b = 0
candidato_c = 0
aux = 0 

qtotal_eleitores = int(input("Informe a quantidade de eleitores"))

while qtotal_eleitores <= 0:
    print("Quantidade de eleitores inválida, precisa ser pelo menos 1")
    qtotal_eleitores = int(input("Informe a quantidade de eleitores: "))

while aux < qtotal_eleitores:  
    # Pedimos o voto do eleitor atual
    voto = input(f"\n[Eleitor {aux + 1}] Informe seu voto ('a', 'b' ou 'c'): ").lower().strip()
    
    # Validação: Garante que o usuário só digite uma das opções válidas
    while voto not in ['a', 'b', 'c']:
        print("Opção inválida! Vote apenas em 'a', 'b' ou 'c'.")
        voto = input(f"[Eleitor {aux + 1}] Informe seu voto ('a', 'b' ou 'c'): ").lower().strip()
    
    # Agora que o voto é garantidamente válido, fazemos a contagem segura
    if voto == 'a':
        candidato_a += 1
    elif voto == 'b':
        candidato_b += 1
    elif voto == 'c':  
        candidato_c += 1
        
    aux += 1    


print(f"\n{' Resultado ':-^35}")
print(f"O candidato A teve {candidato_a} voto(s)")
print(f"O candidato B teve {candidato_b} voto(s)")
print(f"O candidato C teve {candidato_c} voto(s)")
````

#### 10. Faça um programa que calcule o valor total investido por um colecionador em sua coleção de CDs e o valor médio gasto em cada um deles. O usuário deverá informar a quantidade de CDs e o valor para em cada um.

````python
# PLANEJAMENTO:
# 1. Pedir a quantidade de CDs (deve ser maior que 0).
# 2. Criar os contadores e acumuladores (aux para o loop, total_investido para a soma).
# 3. Criar um loop while para rodar até atingir a quantidade de CDs.
# 4. Dentro do loop, pedir o valor de cada CD (validando para não aceitar valores negativos).
# 5. Somar o valor ao total_investido e incrementar o aux.
# 6. Calcular a média e exibir os resultados formatados.

print("--- Gerenciador de Coleção de CDs ---")

# 1. Validação da quantidade de CDs
qtd_cds = int(input("Informe a quantidade de CDs na sua coleção: "))
while qtd_cds <= 0:
    print("Sua coleção precisa ter pelo menos 1 CD!")
    qtd_cds = int(input("Informe a quantidade de CDs na sua coleção: "))

aux = 0
total_investido = 0.0

# 3. Loop principal para coletar os valores
while aux < qtd_cds:
    # Pedimos o valor do CD atual
    valor_cd = float(input(f"Informe o valor pago pelo {aux + 1}º CD: R$ "))
    
    # 4. Validação do valor (não aceita preço negativo, mas aceita 0 caso tenha sido presente)
    while valor_cd < 0:
        print("O valor do CD não pode ser negativo!")
        valor_cd = float(input(f"Informe o valor pago pelo {aux + 1}º CD: R$ "))
        
    # 5. Acumulando o valor total e atualizando o contador
    total_investido += valor_cd
    aux += 1

# 6. Calculando a média
valor_medio = total_investido / qtd_cds

# Exibição dos resultados com estilo pythônico
print(f"\n{' Estatísticas da Coleção ':-^40}")
print(f"Total de CDs cadastrados: {qtd_cds}")
print(f"Valor total investido:   R$ {total_investido:.2f}")
print(f"Valor médio por CD:       R$ {valor_medio:.2f}")
print(f"{'-' * 40}")
````

#### 11. O Sr. Manoel Joaquim possui uma grande loja de artigos de R\\$ 1,99, com cerca de 10 caixas. Para agilizar o cálculo de quanto cada cliente deve pagar ele desenvolveu um tabela que contém o número de itens que o cliente comprou e ao lado o valor da conta. Desta forma a atendente do caixa precisa apenas contar quantos itens o cliente está levando e olhar na tabela de preços. Você foi contratado para desenvolver o programa que monta esta tabela de preços, que conterá os preços de 1 até 50 produtos, conforme o exemplo abaixo:
<pre>
Lojas Quase Dois - Tabela de preços
1 - R$ 1.99
2 - R$ 3.98
...
50 - R$ 99.50
</pre>

````python
# PLANEJAMENTO:
# 1. Criar o cabeçalho da tabela de preços.
# 2. Inicializar o contador de produtos em 1 (já que a tabela começa com 1 item).
# 3. Criar o loop while que vai rodar até o produto número 50.
# 4. Dentro do loop, calcular o preço (quantidade de produtos * 1.99).
# 5. Exibir a linha da tabela formatada.
# 6. Incrementar o contador.

print(f"{' Lojas Quase Dois - Tabela de preços ':-^45}\n")

# Inicializamos o contador com 1 produto
produto = 1

# O loop roda enquanto o número do produto for menor ou igual a 50
while produto <= 50:
    # Calculamos o preço dinamicamente para cada linha
    preco = produto * 1.99
    
    # Exibimos a linha formatada com o número do produto e o preço
    print(f"{produto:02d} - R$ {preco:.2f}")
    
    # Incrementamos o contador para não gerar loop infinito
    produto += 1

print(f"\n{'-' * 45}")
````
