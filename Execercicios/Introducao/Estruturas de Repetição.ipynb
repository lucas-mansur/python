#### 8. Faça um programa que consiga categorizar a idade das equipes de uma empresa. Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da equipe varia entre 0 e 25 (jovem) ,26 e 60 (sênior) e maior que 60 (idosa); e então, dizer se a equipe é jovem, sênior ou idosa, conforme a média calculada.

````python
print("Informe a quantidade de pessoas da equipe para classificar a sua equipe quanto à idade:")

# Validação para garantir que a equipe tenha pelo menos 1 pessoa
qtd = int(input("Informe quantas pessoas tem na sua equipe: "))
while qtd <= 0:
    print("A equipe deve ter pelo menos 1 pessoa.")
    qtd = int(input("Informe quantas pessoas tem na sua equipe: "))

aux = 0
soma_idade = 0

while aux < qtd:
    idade = int(input(f"Informe a idade da {aux + 1}ª pessoa: "))
    
    # Validação da idade: não deixa avançar enquanto não for uma idade realista
    while idade < 0 or idade > 120:
        print("Por favor, insira uma idade válida entre 0 e 120 anos.")
        idade = int(input(f"Informe a idade da {aux + 1}ª pessoa: "))
        
    soma_idade += idade
    aux += 1

media = soma_idade / qtd

if media <= 25:
    classificacao = "jovem"
elif media <= 60:
    classificacao = "sênior"
else:
    classificacao = "idosa"

# Usando o cabeçalho pythônico
print(f"\n{' Resultado ':-^35}")
print(f"Média de idade da equipe: {media:.1f} anos")
print(f"A sua equipe é considerada: {classificacao.upper()}")
````

#### 10. Faça um programa que calcule o valor total investido por um colecionador em sua coleção de CDs e o valor médio gasto em cada um deles. O usuário deverá informar a quantidade de CDs e o valor para em cada um.

````python
# PLANEJAMENTO:
# 1. Pedir a quantidade de CDs (deve ser maior que 0).
# 2. Criar os contadores e acumuladores (aux para o loop, total_investido para a soma).
# 3. Criar um loop while para rodar até atingir a quantidade de CDs.
# 4. Dentro do loop, pedir o valor de cada CD (validando para não aceitar valores negativos).
# 5. Somar o valor ao total_investido e incrementar o aux.
# 6. Calcular a média e exibir os resultados formatados.

print("--- Gerenciador de Coleção de CDs ---")

# 1. Validação da quantidade de CDs
qtd_cds = int(input("Informe a quantidade de CDs na sua coleção: "))
while qtd_cds <= 0:
    print("Sua coleção precisa ter pelo menos 1 CD!")
    qtd_cds = int(input("Informe a quantidade de CDs na sua coleção: "))

aux = 0
total_investido = 0.0

# 3. Loop principal para coletar os valores
while aux < qtd_cds:
    # Pedimos o valor do CD atual
    valor_cd = float(input(f"Informe o valor pago pelo {aux + 1}º CD: R$ "))
    
    # 4. Validação do valor (não aceita preço negativo, mas aceita 0 caso tenha sido presente)
    while valor_cd < 0:
        print("O valor do CD não pode ser negativo!")
        valor_cd = float(input(f"Informe o valor pago pelo {aux + 1}º CD: R$ "))
        
    # 5. Acumulando o valor total e atualizando o contador
    total_investido += valor_cd
    aux += 1

# 6. Calculando a média
valor_medio = total_investido / qtd_cds

# Exibição dos resultados com estilo pythônico
print(f"\n{' Estatísticas da Coleção ':-^40}")
print(f"Total de CDs cadastrados: {qtd_cds}")
print(f"Valor total investido:   R$ {total_investido:.2f}")
print(f"Valor médio por CD:       R$ {valor_medio:.2f}")
print(f"{'-' * 40}")
````

#### 11. O Sr. Manoel Joaquim possui uma grande loja de artigos de R\\$ 1,99, com cerca de 10 caixas. Desta forma a atendente do caixa precisa apenas contar quantos itens o cliente está levando e olhar na tabela de preços. Você foi contratado para desenvolver o programa que monta esta tabela de preços, que conterá os preços de 1 até 50 produtos, conforme o exemplo abaixo:
<pre>
Lojas Quase Dois - Tabela de preços
1 - R$ 1.99
2 - R$ 3.98
...
50 - R$ 99.50
</pre>

````python
# PLANEJAMENTO:
# 1. Criar o cabeçalho da tabela de preços.
# 2. Inicializar o contador de produtos em 1 (já que a tabela começa com 1 item).
# 3. Criar o loop while que vai rodar até o produto número 50.
# 4. Dentro do loop, calcular o preço (quantidade de produtos * 1.99).
# 5. Exibir a linha da tabela formatada.
# 6. Incrementar o contador.

print(f"{' Lojas Quase Dois - Tabela de preços ':-^45}\n")

# Inicializamos o contador com 1 produto
produto = 1

# O loop roda enquanto o número do produto for menor ou igual a 50
while produto <= 50:
    # Calculamos o preço dinamicamente para cada linha
    preco = produto * 1.99
    
    # Exibimos a linha formatada com o número do produto e o preço
    print(f"{produto:02d} - R$ {preco:.2f}")
    
    # Incrementamos o contador para não gerar loop infinito
    produto += 1

print(f"\n{'-' * 45}")
````

#### 12. Um funcionário de uma empresa recebe aumento salarial anualmente: Sabe-se que:
<pre>
Esse funcionário foi contratado em 1995, com salário inicial de R$ 1.000,00;
Em 1996 recebeu aumento de 1,5% sobre seu salário inicial;
A partir de 1997 (inclusive), os aumentos salariais sempre correspondem ao dobro do percentual do ano anterior. Faça um programa que determine o salário desse funcionário em 2003. 
</pre>

````python
# PLANEJAMENTO:
# 1. Definir o ano atual da simulação (começa em 1996, onde ocorre o primeiro aumento).
# 2. Definir o ano limite (2005).
# 3. Definir o salário inicial (R$ 1000.00).
# 4. Definir o percentual inicial de aumento de 1996 (1.5% -> 0.015).
# 5. Aplicar o primeiro aumento de 1996.
# 6. Criar um loop while a partir de 1997 até 2005, dobrando o percentual a cada ano.

salario = 1000.0
ano_atual = 1996
ano_limite = 2005
percentual_aumento = 0.015  # 1.5%

# Aplicando o aumento do ano de 1996
salario += salario * percentual_aumento

# Agora o loop assume a partir de 1997
# O ano_atual passa a ser 1997
ano_atual += 1

while ano_atual <= ano_limite:
    # A partir de 1997, o percentual é o dobro do ano anterior
    percentual_aumento *= 2
    
    # Aplica o novo aumento sobre o salário do ano anterior
    salario += salario * percentual_aumento
    
    # Avança para o próximo ano
    ano_atual += 1

# Exibição do resultado final formatado
print(f"{' Evolução Salarial Concluída ':-^45}")
print(f"Salário final em {ano_limite}: R$ {salario:,.2f}")
print(f"{'-' * 45}")
````

Faça um programa que leia o código dos itens pedidos e as quantidades desejadas. Calcule e mostre o valor a ser pago por item (preço * quantidade) e o total geral do pedido. Considere que o cliente deve informar quando o pedido deve ser encerrado.

````python
# PLANEJAMENTO:
# 1. Exibir o cardápio para o usuário.
# 2. Criar a variável acumuladora total_geral = 0.
# 3. Iniciar o loop 'while True'.
# 4. Pedir o código do produto (e avisar que 0 ou outro número específico encerra o pedido).
# 5. Se o código for o de encerramento, quebrar o loop (break).
# 6. Se for um código válido, pedir a quantidade e calcular o valor daquele item.
# 7. Somar o valor do item ao total_geral.
# 8. Ao final do loop, exibir a conta fechada.

print(f"{' LANCHONETE PYTHON ':=^40}")
print("""
Especificação    Código   Preço
Cachorro Quente  100      R$ 1,20
Bauru Simples    101      R$ 1,30
Bauru com ovo    102      R$ 1,50
Hambúrguer       103      R$ 1,20
Cheeseburguer    104      R$ 1,30
Refrigerante     105      R$ 1,00

Digite 0 a qualquer momento para encerrar o pedido.
""")
print("=" * 40)

total_geral = 0.0

while True:
    codigo = int(input("\nInforme o código do item desejado: "))
    
    # Condição de parada do loop
    if codigo == 0:
        print("\nFechando a comanda...")
        break
        
    # Validação do código digitado
    if codigo not in [100, 101, 102, 103, 104, 105]:
        print("Código inválido! Por favor, tente novamente.")
        continue # Volta para o início do loop sem pedir a quantidade
        
    quantidade = int(input("Informe a quantidade desejada: "))
    while quantidade <= 0:
        print("A quantidade deve ser maior que zero.")
        quantidade = int(input("Informe a quantidade desejada: "))
        
    # Processamento do item com base no código escolhido
    if codigo == 100:
        item = "Cachorro Quente"
        preco_unitario = 1.20
    elif codigo == 101:
        item = "Bauru Simples"
        preco_unitario = 1.30
    elif codigo == 102:
        item = "Bauru com ovo"
        preco_unitario = 1.50
    elif codigo == 103:
        item = "Hambúrguer"
        preco_unitario = 1.20
    elif codigo == 104:
        item = "Cheeseburguer"
        preco_unitario = 1.30
    elif codigo == 105:
        item = "Refrigerante"
        preco_unitario = 1.00

    # Cálculo do item atual
    valor_item = preco_unitario * quantidade
    total_geral += valor_item
    
    # Mostra o subtotal do item atual lançado
    print(f"-> {quantidade}x {item}: R$ {valor_item:.2f}")

# Cupom Fiscal final
print(f"\n{' NOTA FISCAL ':-^40}")
print(f"TOTAL GERAL A PAGAR: R$ {total_geral:.2f}")
print(f"{'-' * 40}")
````

````python
# PLANEJAMENTO COM DICIONÁRIO (DICT):
# 1. Criar o dicionário 'cardapio' onde a CHAVE é o código e o VALOR são os dados do produto (nome e preço).
# 2. Exibir o cardápio na tela e inicializar o 'total_geral' em 0.0.
# 3. Iniciar o loop 'while True' (para rodar até o usuário decidir parar).
# 4. Pedir o código do item. Se for 0, quebrar o loop com 'break'.
# 5. MUDANÇA COM DICT: Em vez de testar com vários 'if/elif', verificar se o código NÃO está no dicionário usando 'not in'.
# 6. Se o código for inválido, exibir erro e usar 'continue' para reiniciar o loop.
# 7. Pedir a quantidade e usar um 'while' de validação para garantir que seja maior que zero.
# 8. MUDANÇA COM DICT: Buscar o nome e o preço direto do dicionário usando a chave 'codigo', eliminando os blocos de decisão.
# 9. Calcular o valor do item, somar ao 'total_geral' e exibir o subtotal.
# 10. Fora do loop, exibir o cupom fiscal com o total a pagar.

# 1. Criação da estrutura de dados
cardapio = {
    100: {"nome": "Cachorro Quente", "preco": 1.20},
    101: {"nome": "Bauru Simples",   "preco": 1.30},
    102: {"nome": "Bauru com ovo",   "preco": 1.50},
    103: {"nome": "Hambúrguer",      "preco": 1.20},
    104: {"nome": "Cheeseburguer",   "preco": 1.30},
    105: {"nome": "Refrigerante",    "preco": 1.00}
}

# 2. Inicialização do sistema
print(f"{' LANCHONETE PYTHON ':=^45}")
print("""
Especificação    Código   Preço
Cachorro Quente  100      R$ 1,20
Bauru Simples    101      R$ 1,30
Bauru com ovo    102      R$ 1,50
Hambúrguer       103      R$ 1,20
Cheeseburguer    104      R$ 1,30
Refrigerante     105      R$ 1,00

Digite 0 para encerrar o pedido.
""")
print("=" * 45)

total_geral = 0.0

# 3. Loop principal de vendas
while True:
    # 4. Leitura do código e condição de saída
    codigo = int(input("\nInforme o código do item desejado: "))
    
    if codigo == 0:
        print("\nFechando a comanda...")
        break
        
    # 5 e 6. Validação rápida usando as chaves do dicionário
    if codigo not in cardapio:
        print("Código inválido! Por favor, tente novamente.")
        continue 
        
    # 7. Validação da quantidade (garantindo valor positivo)
    quantidade = int(input("Informe a quantidade desejada: "))
    while quantidade <= 0:
        print("A quantidade deve ser maior que zero.")
        quantidade = int(input("Informe a quantidade desejada: "))
        
    # 8. Mágica do Dict: Acessando os dados diretamente, sem nenhum if/elif
    item_escolhido = cardapio[codigo]
    nome = item_escolhido["nome"]
    preco_unitario = item_escolhido["preco"]

    # 9. Cálculo e atualização dos acumuladores
    valor_item = preco_unitario * quantidade
    total_geral += valor_item
    
    print(f"-> {quantidade}x {nome}: R$ {valor_item:.2f}")

# 10. Emissão do resultado final
print(f"\n{' NOTA FISCAL ':-^45}")
print(f"TOTAL GERAL A PAGAR: R$ {total_geral:.2f}")
print(f"{'-' * 45}")
````

````python
# PLANEJAMENTO COM LISTA (ARRAY):
# 1. Criar uma lista com 24 posições zeradas (usamos 24 para a posição 23 existir direto).
# 2. Inicializar o 'total_votos' em 0.
# 3. Iniciar o loop 'while True' para ler os votos.
# 4. Se o voto for 0, quebrar o loop com 'break'.
# 5. Validar se o voto está entre 1 e 23. Se não estiver, mostrar aviso e usar 'continue'.
# 6. Se for válido, somar 1 no 'total_votos' e somar 1 na posição da lista correspondente ao número do jogador.
# 7. Criar um loop 'while' (de 1 até 23) para percorrer a lista, calcular os percentuais e exibir quem teve voto.
# 8. Dentro desse mesmo loop, atualizar quem está sendo o melhor jogador.
# 9. Exibir o resultado final do melhor jogador.

print("Enquete: Quem foi o melhor jogador?\n")

# 1. Cria uma lista com 24 zeros. 
# O índice 0 vai ficar vazio, e usaremos os índices de 1 a 23 para os jogadores.
votos_jogadores = [0] * 24
total_votos = 0

# 3. Loop principal de votação
while True:
    voto = int(input("Número do jogador (0=fim): "))
    
    # 4. Condição de parada
    if voto == 0:
        break
        
    # 5. Validação clássica
    if voto < 1 or voto > 23:
        print("Informe um valor entre 1 e 23 ou 0 para sair!")
        continue
        
    # 6. Computação do voto na lista usando o próprio número digitado como posição
    votos_jogadores[voto] += 1
    total_votos += 1

print("\nResultado da votação:")
print(f"Foram computados {total_votos} votos.\n")
print(f"{'Jogador':<9}{'Votos':<15}%")

# Variáveis para rastrear o melhor jogador
melhor_jogador = 0
votos_melhor = -1

# 7. Loop para varrer a lista do jogador 1 até o 23
camisa = 1
while camisa <= 23:
    votos_atual = votos_jogadores[camisa]
    
    # Se o jogador recebeu votos, calcula e exibe
    if votos_atual > 0:
        percentual = (votos_atual / total_votos) * 100
        print(f"{camisa:<9}{votos_atual:<15}{percentual:.1f}%")
        
        # 8. Lógica para descobrir o melhor jogador
        if votos_atual > votos_melhor:
            votos_melhor = votos_atual
            melhor_jogador = camisa
            
    camisa += 1

# 9. Exibição do campeão
if total_votos > 0:
    percentual_melhor = (votos_melhor / total_votos) * 100
    print(f"\nO melhor jogador foi o número {melhor_jogador}, com {votos_melhor} votos, correspondendo a {percentual_melhor:.1f}% do total de votos.")
````
