# Listas importantes a partir do Dicionário

### Métodos importantes:

.keys() -> uma "lista" com todas as chaves do dicionario

.values() -> uma "lista" com todos os valores do dicionario

Obs: Se o dicionario for modificado, automaticamente essas variáveis são modificadas, mesmo tendo sido criadas anteriormente

```python
chaves = vendas_tecnologia.keys()
valores = vendas_tecnologia.values()
print(chaves)
print(valores)
```

```python
vendas_tecnologia['liraphone'] = 10
print(chaves)
print(valores)
print(list(chaves))
print(list(valores))
```

```python
# PLANEJAMENTO DO NOTEBOOK DE ESTUDOS:
# 1. Definir o dicionário base 'vendas_tecnologia'.
# 2. Resolver o Desafio 1 (Produtos > 5000 unidades) usando a Forma 1 (Chave direta).
# 3. Resolver o Desafio 1 usando a Forma 2 (.items() com desempacotamento de tuplas).
# 4. Demonstrar o comportamento dinâmico de .keys() e .values() ao adicionar 'liraphone'.
# 5. Converter as dict_listas em listas reais usando list().
# 6. Resolver o Desafio 2 (Ordenação Alfabética): extrair chaves, ordenar com .sort() e exibir os dados agrupados.

# Configuração inicial do dicionário de tecnologias
vendas_tecnologia = {
    'notebook asus': 2450, 
    'iphone': 15000, 
    'samsung galaxy': 12000, 
    'tv samsung': 10000, 
    'ps5': 14300, 
    'tablet': 1720, 
    'notebook dell': 17000, 
    'ipad': 1000, 
    'tv philco': 2500, 
    'notebook hp': 1000
}

print(f"{' FILTRO: MAIS DE 5000 UNIDADES ':=^50}\n")

# ------------------------------------------------------------------------------
# FORMA 1 -> usando apenas o dicionario e as chaves
# ------------------------------------------------------------------------------
print("Forma 1 (Acesso por chave direta):")
for chave in vendas_tecnologia:
    if vendas_tecnologia[chave] > 5000:
        print('  - {}: {} unidades'.format(chave, vendas_tecnologia[chave]))

print()

# ------------------------------------------------------------------------------
# FORMA 2 -> usando o dicionario.items()
# ------------------------------------------------------------------------------
print("Forma 2 (Usando .items() e Desempacotamento de Tuplas):")
# O .items() transforma o par em uma tupla e o for já desempacota nas variáveis
for produto, unidades in vendas_tecnologia.items():
    if unidades > 5000:
        print('  - {}: {} unidades'.format(produto, unidades))


print(f"\n{' LISTAS E MÉTODOS .KEYS() E .VALUES() ':=^50}\n")
```

```python
print(f"\n{' LISTAS E MÉTODOS .KEYS() E .VALUES() ':=^50}\n")

# Criando as visões dinâmicas (dict_lists)
chaves = vendas_tecnologia.keys()
valores = vendas_tecnologia.values()

print("Visão inicial das chaves:")
print(chaves)
print("Visão inicial os valores:")
print(valores)
print("-" * 40)
```

```python
# Adicionando um novo item para testar a modificação em tempo real (Dinamismo)
vendas_tecnologia['liraphone'] = 10

print("Visão atualizada automaticamente (Sem rodar o .keys() de novo):")
print(chaves)
print(valores)
print("-" * 40)

# Convertendo as dict_listas em listas normais usando a função list()
print("Convertidas em listas reais do Python:")
print(list(chaves))
print(list(valores))
```

```python
print(f"\n{' EXIBIÇÃO PADRÃO (SEM ORDEM) ':-^50}")
for chave in vendas_tecnologia:
     print('{}: {} unidades'.format(chave, vendas_tecnologia[chave]))

print('=' * 50)
print(f"{' EXIBIÇÃO ORGANIZADA ALFABETICAMENTE ':-^50}")

# 1. Transformamos a visão dict_keys em uma lista comum do Python
lista_chaves = list(vendas_tecnologia.keys())

# 2. Usamos o método .sort() das listas para ordenar as strings alfabeticamente
lista_chaves.sort()

# 3. Percorremos a lista já organizada e buscamos os valores de volta no dict
for chave in lista_chaves:
     print('{}: {} unidades'.format(chave.title(), vendas_tecnologia[chave]))

print('=' * 50)
```
