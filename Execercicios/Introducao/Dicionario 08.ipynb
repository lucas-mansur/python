```python
# PLANEJAMENTO DO NOTEBOOK:
# 1. Definir as listas paralelas de 'produtos' e 'vendas'.
# 2. Responder ao Desafio (Vendas do iPad) usando a FORMA 1 (Apenas listas e o método .index()).
# 3. Criar o dicionário combinando as duas listas através da função zip().
# 4. Responder ao Desafio usando a FORMA 2 (Busca direta por chave no novo dicionário).
# 5. Exibir os resultados de ambas as formas para comparação de performance e clareza.

# Listas paralelas originais
produtos = ['iphone', 'samsung galaxy', 'tv samsung', 'ps5', 'tablet', 'ipad', 'tv philco', 'notebook hp', 'notebook dell', 'notebook asus']
vendas = [15000, 12000, 10000, 14300, 1720, 1000, 2500, 1000, 17000, 2450]

print(f"{' RESOLUÇÃO DO DESAFIO ':=^50}\n")

# ------------------------------------------------------------------------------
# Fazendo por listas (Forma 1)
# ------------------------------------------------------------------------------
# Passo 1: Descobrir em qual índice a string 'ipad' está na lista de produtos
indice_ipad = produtos.index('ipad')

# Passo 2: Usar o mesmo índice para buscar o valor correspondente na lista de vendas
vendas_ipad_lista = vendas[indice_ipad]

print("Forma 1 (Usando apenas Listas):")
print(f"O produto 'ipad' está no índice {indice_ipad}.")
print(f"Total vendido: {vendas_ipad_lista} unidades.\n")


# ------------------------------------------------------------------------------
# Fazendo por dicionário (Forma 2)
# ------------------------------------------------------------------------------
# Passo 1: "Zipar" as duas listas paralelas em formato de tuplas
# Passo 2: Converter o resultado diretamente para a estrutura de dicionário
dicionario_vendas = dict(zip(produtos, vendas))

# Passo 3: Buscar o valor direto pela chave
vendas_ipad_dict = dicionario_vendas['ipad']

print("Forma 2 (Usando Dicionário + Zip):")
print(f"Dicionário criado: {dicionario_vendas}")
print(f"Total vendido de iPad (Busca por chave): {vendas_ipad_dict} unidades.")

print(f"\n{'=' * 50}")
```

🧠 Conhecimento Agregado para o seu resumo teórico:
Note a diferença gritante de lógica entre as duas formas:

Pelas Listas: Você precisa obrigar o Python a varrer a lista de produtos procurando pelo 'ipad' com o .index(), para só depois ir na outra lista pegar o valor. Se as listas saírem de sincronia (um produto for deletado de uma lista mas não da outra), seu código trará o valor errado!

Pelo Dicionário com zip: O zip amarra permanentemente o produto ao seu respectivo valor na hora da criação. Depois disso, a busca por dicionario_vendas['ipad'] é instantânea e imune a erros de dessincronização.
