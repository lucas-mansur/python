- Qual o total de notebooks vendidos?

```python
# PLANEJAMENTO DO FILTRO DE NOTEBOOKS:
# 1. Definir o dicionário base 'vendas_tecnologia'.
# 2. Criar uma variável acumuladora 'total_notebooks' inicializada com 0.
# 3. Criar o loop 'for' padrão para percorrer as chaves do dicionário ('for produto in vendas_tecnologia:').
# 4. Dentro do loop, usar uma estrutura de decisão 'if' para verificar se a palavra 'notebook' está contida na chave atual.
# 5. Se estiver, "pescar" o valor de vendas usando 'vendas_tecnologia[produto]' e somar ao acumulador.
# 6. Exibir o total acumulado de forma organizada.

# 1. Base de dados
vendas_tecnologia = {
    'iphone': 15000, 'samsung galaxy': 12000, 'tv samsung': 10000, 
    'ps5': 14300, 'tablet': 1720, 'ipad': 1000, 'tv philco': 2500, 
    'notebook hp': 1000, 'notebook dell': 17000, 'notebook asus': 2450
}

# ------------------------------------------------------------------------------
# Demonstrando o funcionamento básico do FOR
# ------------------------------------------------------------------------------
print(f"{' DEMONSTRAÇÃO DO FOR PADRÃO ':=^45}\n")

# Por padrão, a variável 'produto' recebe apenas o TEXTO da chave
for produto in vendas_tecnologia:
    vendas = vendas_tecnologia[produto] # Buscando o valor correspondente à chave
    print(f"Produto: {produto:<15} | Vendeu: {vendas} unidades")


# ------------------------------------------------------------------------------
# Respondendo à pergunta: Qual o total de notebooks vendidos?
# ------------------------------------------------------------------------------
print(f"\n{' ANÁLISE DE VENDAS DE NOTEBOOKS ':=^45}\n")

total_notebooks = 0

# Percorrendo o dicionário para filtrar e somar
for produto in vendas_tecnologia:
    # Verificando se a palavra 'notebook' faz parte do nome do produto
    if 'notebook' in produto:
        print(f"-> Encontrado: {produto.title()} ({vendas_tecnologia[produto]} un.)")
        # Somando as unidades ao nosso totalizador
        total_notebooks += vendas_tecnologia[produto]

print("-" * 45)
print(f"TOTAL GERAL DE NOTEBOOKS VENDIDOS: {total_notebooks} unidades.")
print("-" * 45)
```
