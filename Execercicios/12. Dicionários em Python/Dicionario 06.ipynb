# Listas importantes a partir do Dicionário

### Métodos importantes:

.keys() -> uma "lista" com todas as chaves do dicionario

.values() -> uma "lista" com todos os valores do dicionario

Obs: Se o dicionario for modificado, automaticamente essas variáveis são modificadas, mesmo tendo sido criadas anteriormente

```python
chaves = vendas_tecnologia.keys()
valores = vendas_tecnologia.values()
print(chaves)
print(valores)
```

```python
vendas_tecnologia['liraphone'] = 10
print(chaves)
print(valores)
print(list(chaves))
print(list(valores))
```

```python
# PLANEJAMENTO DE RESOLUÇÃO DO NOTEBOOK:
# 1. Definir o dicionário base 'vendas_tecnologia'.
# 2. Desafio 1 (Produtos > 5000 unidades) - Forma 1: Percorrer chaves e buscar valores com dict[chave].
# 3. Desafio 1 (Produtos > 5000 unidades) - Forma 2: Usar .items() com desempacotamento de tuplas no for.
# 4. Demonstrar o comportamento dinâmico de .keys() e .values() ao incluir o 'liraphone'.
# 5. Desafio 2 (Organização Alfabética): Converter as chaves em uma lista real usando list(), ordená-la com sorted() ou .sort() e exibir os dados agrupados.


# ==============================================================================
# CONFIGURAÇÃO INICIAL DO NOTEBOOK
# ==============================================================================
vendas_tecnologia = {
    'notebook asus': 2450, 'iphone': 15000, 'samsung galaxy': 12000, 
    'tv samsung': 10000, 'ps5': 14300, 'tablet': 1720, 
    'notebook dell': 17000, 'ipad': 1000, 'tv philco': 2500, 'notebook hp': 1000
}

# ------------------------------------------------------------------------------
# QUAIS PRODUTOS VENDERAM MAIS DE 5000 UNIDADES?
# ------------------------------------------------------------------------------
print(f"{' FILTRO: MAIS DE 5000 UNIDADES ':=^50}\n")

# FORMA 1 -> Usando apenas o dicionário e as chaves
print("Forma 1 (Acesso por chave direta):")
for chave in vendas_tecnologia:
    if vendas_tecnologia[chave] > 5000:
        print(f"  - {chave:<15}: {vendas_tecnologia[chave]} unidades")

print()

# FORMA 2 -> Usando o dicionário.items() (Mais Pythônica!)
# O .items() entrega a tupla (chave, valor) pronta e o for já desempacota
print("Forma 2 (Usando .items() e Desempacotamento):")
for produto, unidades in vendas_tecnologia.items():
    if unidades > 5000:
        print(f"  - {produto:<15}: {unidades} unidades")


# ------------------------------------------------------------------------------
# LISTAS IMPORTANTES A PARTIR DO DICIONÁRIO (COMPORTAMENTO DINÂMICO)
# ------------------------------------------------------------------------------
print(f"\n{' TESTE DE DINAMISMO (.keys() / .values()) ':=^50}\n")

chaves = vendas_tecnologia.keys()
valores = vendas_tecnologia.values()

print("Antes da modificação:")
print(f"Chaves:  {chaves}")
print(f"Valores: {valores}\n")

# Modificando o dicionário...
vendas_tecnologia['liraphone'] = 10

print("Depois da modificação (O 'espelho' atualiza sozinho):")
print(f"Chaves:  {chaves}")
print(f"Valores: {valores}\n")

print("Convertendo os espelhos em listas normais usando list():")
print(f"Lista de Chaves:  {list(chaves)}")
print(f"Lista de Valores: {list(valores)}")


# ------------------------------------------------------------------------------
# ORGANIZAÇÃO ALFABÉTICA (ORDENANDO AS CHAVES)
# ------------------------------------------------------------------------------
print(f"\n{' EXIBIÇÃO ORDENADA ALFABETICAMENTE ':=^50}\n")

# 1. Transformamos a visão dict_keys em uma lista comum do Python
lista_chaves = list(vendas_tecnologia.keys())

# 2. Usamos o método .sort() para ordenar a lista de strings em ordem alfabética
lista_chaves.sort()

# 3. Agora percorremos a lista já ordenada e buscamos o valor de cada uma no dict
for chave in lista_chaves:
    print(f"{chave.title():<20}: {vendas_tecnologia[chave]} unidades")

print(f"\n{'=' * 50}")
```
