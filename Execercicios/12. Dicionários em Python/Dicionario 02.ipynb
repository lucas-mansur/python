# Não confie na ordem dos dicionários, use as chaves

### Python Versões Antigas x Versões Novas

- Dicionários eram "sem ordem". Atualmente tem ordem, mas o certo é usar as chaves
- 2 formas de pegar um valor:
    - dicionario[chave]
    - .get(chave)

```python
mais_vendidos = {'tecnologia': 'iphone', 'refrigeracao': 'ar consul 12000 btu', 'livros': 'o alquimista', 'eletrodoméstico': 'geladeira', 'lazer': 'prancha surf'}

vendas_tecnologia = {'iphone': 15000, 'samsung galaxy': 12000, 'tv samsung': 10000, 'ps5': 14300, 'tablet': 1720, 'ipad': 1000, 'tv philco': 2500, 'notebook hp': 1000, 'notebook dell': 17000, 'notebook asus': 2450}
```

- Qual foi o item mais vendido nas categorias 'livros' e 'lazer'?
- Quanto foi vendido de 'notebook asus' e de 'ipad'?

```python
# PLANEJAMENTO DE ACESSO A DICT:
# 1. Definir os dicionários base 'mais_vendidos' e 'vendas_tecnologia'.
# 2. PARTE 1 (Abordagem por Chave Direta):
#    - Buscar os itens mais vendidos de 'livros' e 'lazer' usando dicionario[chave].
#    - Exibir os resultados.
# 3. PARTE 2 (Abordagem pelo Método .get()):
#    - Buscar as vendas de 'notebook asus' e 'ipad' usando dicionario.get(chave).
#    - Exibir os resultados.

# 1. Base de dados
mais_vendidos = {
    'tecnologia': 'iphone', 
    'refrigeracao': 'ar consul 12000 btu', 
    'livros': 'o alquimista', 
    'eletrodoméstico': 'geladeira', 
    'lazer': 'prancha surf'
}

vendas_tecnologia = {
    'iphone': 15000, 
    'samsung galaxy': 12000, 
    'tv samsung': 10000, 
    'ps5': 14300, 
    'tablet': 1720, 
    'ipad': 1000, 
    'tv philco': 2500, 
    'notebook hp': 1000, 
    'notebook dell': 17000, 
    'notebook asus': 2450
}

# ------------------------------------------------------------------------------
# PARTE 1: Respondendo com a chave direta (Igual à aula anterior)
# ------------------------------------------------------------------------------
print(f"{' ABORDAGEM 1: CHAVE DIRETA [CHAVE] ':=^50}\n")

# Se a chave não existir aqui, o programa TRAVA com KeyError
livro = mais_vendidos['livros']
lazer = mais_vendidos['lazer']

print(f"Mais vendido em Livros: {livro.title()}")
print(f"Mais vendido em Lazer:  {lazer.title()}")


# ------------------------------------------------------------------------------
# PARTE 2: Respondendo com o método .get()
# ------------------------------------------------------------------------------
print(f"\n{' ABORDAGEM 2: MÉTODO .GET(CHAVE) ':=^50}\n")

# Se a chave não existir aqui, ele retorna None (ou o valor padrão que você escolher) de forma segura
vendas_asus = vendas_tecnologia.get('notebook asus')
vendas_ipad = vendas_tecnologia.get('ipad')

# Exemplo bônus de segurança: procurando um produto que NÃO existe no estoque
vendas_macbook = vendas_tecnologia.get('macbook', 0) # Se não achar 'macbook', assume 0 automaticamente

print(f"Vendas de Notebook Asus: {vendas_asus} unidades")
print(f"Vendas de iPad:          {vendas_ipad} unidades")
print(f"Vendas de Macbook:       {vendas_macbook} unidades (Produto não localizado)")

print(f"\n{'-' * 50}")
```

### Verificar se item está no dicionário:

- if
- .get(chave) = None

Se tentarmos procurar as vendas de "copo" na lista de vendas tecnologia, o que acontece?

```python
# PLANEJAMENTO DE VERIFICAÇÃO EM DICIONÁRIOS:
# 1. Definir o dicionário 'vendas_tecnologia' com os produtos existentes.
# 2. Definir o produto a ser buscado ('copo').
# 3. ABORDAGEM 1 (Usando 'if' com o operador 'in'):
#    - Verificar se a string 'copo' está contida nas chaves do dicionário.
#    - Se estiver, exibe o valor. Caso contrário, exibe uma mensagem de aviso.
# 4. ABORDAGEM 2 (Usando '.get()' comparando com 'None'):
#    - Fazer a busca com .get(). Se o Python não encontrar a chave, ele retornará 'None'.
#    - Usar um 'if' para checar se o resultado foi 'None' e tratar o erro de forma limpa.

# 1. Base de dados
vendas_tecnologia = {
    'iphone': 15000, 'samsung galaxy': 12000, 'tv samsung': 10000, 
    'ps5': 14300, 'tablet': 1720, 'ipad': 1000, 'notebook asus': 2450
}

produto_procurado = 'copo'

# ------------------------------------------------------------------------------
# ESTRATÉGIA 1: Usando o "if" tradicional com "in"
# ------------------------------------------------------------------------------
print(f"{' ESTRATÉGIA 1: OPERADOR IN ':=^45}\n")

# O operador 'in' varre as chaves do dicionário de forma ultra rápida
if produto_procurado in vendas_tecnologia:
    print(f"Vendas de {produto_procurado}: {vendas_tecnologia[produto_procurado]} unidades.")
else:
    print(f"Aviso: O produto '{produto_procurado}' não existe no dicionário!")


# ------------------------------------------------------------------------------
# ESTRATÉGIA 2: Usando o ".get(chave) is None"
# ------------------------------------------------------------------------------
print(f"\n{' ESTRATÉGIA 2: MÉTODO .GET() ':=^45}\n")

busca_produto = vendas_tecnologia.get(produto_procurado)

# Em Python, a boa prática é usar 'is None' ou 'is not None' para checar o tipo None
if busca_produto is None:
    print(f"Aviso: .get() retornou None. '{produto_procurado}' não foi localizado.")
else:
    print(f"Vendas de {produto_procurado}: {busca_produto} unidades.")

print(f"\n{'-' * 45}")

```
