### Exercício 1
Crie um sistema de consulta de preços
Seu sistema deve:
- Pedir para o usuário o nome de um produto
- Caso o produto exista na lista de produtos, o programa deve retornar o preço do produto como resposta
       - Ex: O produto celular custa R$1500
- Caso o produto não exista na lista de produtos, o programa deve printar uma mensagem para o usuário tentar novamente


```python
precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 2000}
```

```python
# PLANEJAMENTO DO EXERCÍCIO 1:
# 1. Definir o dicionário base 'precos'.
# 2. Capturar o nome do produto digitado pelo usuário usando input() e aplicar .lower() para padronizar.
# 3. Usar a estrutura 'if produto in precos:' para validar a existência da chave.
# 4. Se existir, buscar o preço e exibir formatado em R$.
# 5. Se não existir, exibir uma mensagem de erro pedindo para tentar novamente.

precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 2000}

# Capturando a entrada do usuário e tratando o texto para minúsculo
produto_procurado = input("Digite o nome do produto para consulta: ").lower()

# Verificando se o produto está nas chaves do dicionário
if produto_procurado in precos:
    preco_produto = precos[produto_procurado]
    print(f"O produto '{produto_procurado.title()}' custa R$ {preco_produto:,.2f}")
else:
    print(f"O produto '{produto_procurado}' não foi localizado. Por favor, tente novamente.")
```


### Exercício 2
Agora edite o programa anterior para fazer com que, caso não exista o produto, o programa pergunte se o usuário quer cadastrar o produto
<br>Se ele responder sim, o programa deve pedir o nome do produto e o preco do produto e cadastrar no dicionário de preços
<br>Em seguida do cadastro bem sucedido, o programa deve printar o dicionário de precos atualizado


```python
# PLANEJAMENTO DO EXERCÍCIO 2:
# 1. Copiar a estrutura do Exercício 1.
# 2. No bloco 'else' (quando o produto não existe), fazer uma nova pergunta usando input(): "Deseja cadastrar?".
# 3. Se a resposta for "sim", solicitar o preço do produto usando float(input()).
# 4. Inserir o novo par chave-valor no dicionário de preços.
# 5. Exibir o dicionário de preços atualizado.

precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 2000}

produto_procurado = input("Digite o nome do produto para consulta: ").lower()

if produto_procurado in precos:
    preco_produto = precos[produto_procurado]
    print(f"O produto '{produto_procurado.title()}' custa R$ {preco_produto:,.2f}")
else:
    print(f"O produto '{produto_procurado}' não existe na lista.")
    
    # Pergunta bônus de cadastro
    opcao = input("Deseja cadastrar esse produto? (sim/não): ").lower()
    
    if opcao == "sim":
        # Captura o preço e adiciona diretamente no dicionário
        novo_preco = float(input(f"Digite o preço para o produto '{produto_procurado}': R$ "))
        precos[produto_procurado] = novo_preco
        
        print("\nCadastro realizado com sucesso!")
        print(f"Dicionário de preços atualizado: {precos}")
    else:
        print("Consulta encerrada.")
```

### Exercício 3
Dada a lista de preços de produtos, uma loja resolveu fazer um reajuste nos preços dos produtos. 
calcule o novo valor dos produtos com base nas seguintes regras:
- Preços até 1.000 vão ter um reajuste de 10% (ou seja, o novo preço será 110% do preço atual)
- Preços até maiores que 1.000 até 2.000 vão ter reajuste de 15%
- Preços acima de 2.000 vão ter reajuste de 20%

```python
precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 3000}
```

```python
# PLANEJAMENTO DO EXERCÍCIO 3:
# 1. Definir o dicionário base 'precos'.
# 2. Criar um loop 'for' percorrendo o dicionário com o método .items() para ter acesso ao produto e ao preço atual.
# 3. Aplicar as regras de faixas de preço usando if/elif/else:
#    - Preço <= 1000: reajuste de 10% (* 1.10)
#    - Preço <= 2000: reajuste de 15% (* 1.15)
#    - Preço > 2000: reajuste de 20% (* 1.20)
# 4. Modificar o valor da chave atual diretamente no dicionário.
# 5. Exibir a lista final de preços atualizados.

precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 3000}

print(f"{' REAJUSTE DE PREÇOS EM LOTE ':=^50}\n")

# Varrendo o dicionário para aplicar as regras de negócio
for produto, preco_atual in precos.items():
    if preco_atual <= 1000:
        precos[produto] = preco_atual * 1.10
    elif preco_atual <= 2000:
        precos[produto] = preco_atual * 1.15
    else:
        precos[produto] = preco_atual * 1.20

# Exibindo os preços após a alteração
for produto, novo_preco in precos.items():
    print(f"Produto: {produto.title():<15} | Novo Preço: R$ {novo_preco:,.2f}")
```

### Exercício 4
 - Edite o programa antigo para ter os 2 dicionários, o de preços originais e o de novos preços
 - Em seguida calcule o valor total de reajuste em R$ que teve entre a lista de produtos original e a lista final


```python
# PLANEJAMENTO DO EXERCÍCIO 4:
# 1. Definir o dicionário original 'precos'.
# 2. Criar um dicionário vazio 'novos_precos = {}' para não apagar o histórico original.
# 3. Inicializar uma variável 'total_original = 0' e 'total_novo = 0' para calcular a diferença.
# 4. Percorrer o dicionário original aplicando as mesmas regras de cálculo do Exercício 3, salvando o resultado no dicionário 'novos_precos'.
# 5. Somar os valores totais de ambos os dicionários usando a função sum() aplicada a .values().
# 6. Calcular a diferença em R$ e exibir os resultados na tela.

precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 3000}
novos_precos = {}

# Preenchendo o novo dicionário com base nas regras do anterior
for produto, preco_atual in precos.items():
    if preco_atual <= 1000:
        novos_precos[produto] = preco_atual * 1.10
    elif preco_atual <= 2000:
        novos_precos[produto] = preco_atual * 1.15
    else:
        novos_precos[produto] = preco_atual * 1.20

# Calculando o impacto financeiro total usando sum() nos valores (.values())
custo_total_original = sum(precos.values())
custo_total_novo = sum(novos_precos.values())
diferenca_reajuste = custo_total_novo - custo_total_original

print(f"{' IMPACTO FINANCEIRO DO REAJUSTE ':=^50}\n")
print(f"Total da tabela antiga: R$ {custo_total_original:,.2f}")
print(f"Total da tabela nova:   R$ {custo_total_novo:,.2f}")
print(f"Custo total do reajuste para a empresa: R$ {diferenca_reajuste:,.2f}")
print("-" * 50)
```


### Exercício 5
- Uma empresa está analisando os resultados de vendas do 1º semestre de 2022 e 2023
- Qual foi o % de crescimento de cada mês de 2023 em relação a 2022?
- Depois de calcular isso, calcule o valor total de crescimento de 2023 em relação a 2022

```python
vendas_22 = {"jan": 15000, "fev": 15500, "mar": 14000, "abr": 16600, "mai": 16300, "jun": 17000}
vendas_23 = {"jan": 17000, "fev": 15000, "mar": 17500, "abr": 16900, "mai": 16000, "jun": 18500}
```

```python
# PLANEJAMENTO DO EXERCÍCIO 5:
# 1. Definir os dicionários 'vendas_22' e 'vendas_23'.
# 2. Criar um loop 'for' para percorrer as chaves de um dos dicionários (ex: 'for mes in vendas_22:').
# 3. Buscar os valores correspondentes daquele mês em ambos os anos.
# 4. Calcular o percentual de crescimento mensal usando a fórmula: (vendas_23 / vendas_22 - 1) * 100.
# 5. Exibir o percentual de cada mês na tela.
# 6. Fora do loop, calcular a soma total de vendas de 2022 e 2023 usando sum().
# 7. Calcular o indicador consolidado do semestre e exibir.

vendas_22 = {"jan": 15000, "fev": 15500, "mar": 14000, "abr": 16600, "mai": 16300, "jun": 17000}
vendas_23 = {"jan": 17000, "fev": 15000, "mar": 17500, "abr": 16900, "mai": 16000, "jun": 18500}

print(f"{' ANÁLISE DE CRESCIMENTO MENSAL ':=^50}\n")

# Percorrendo os meses para calcular o percentual de cada um
for mes in vendas_22:
    venda_antiga = vendas_22[mes]
    venda_nova = vendas_23[mes]
    
    crescimento_mes = (venda_nova / venda_antiga - 1) * 100
    print(f"Mês: {mes.upper()} | 2022: {venda_antiga:<6} | 2023: {venda_nova:<6} | Crescimento: {crescimento_mes:+.1f}%")

# Calculando os totais gerais consolidados do semestre
total_2022 = sum(vendas_22.values())
total_2023 = sum(vendas_23.values())
crescimento_total_valor = total_2023 - total_2022
crescimento_total_percentual = (total_2023 / total_2022 - 1) * 100

print("\n" + "=" * 50)
print(f"{' CONSOLIDADO DO 1º SEMESTRE ':^50}")
print("=" * 50)
print(f"Faturamento Total 2022: R$ {total_2022:,.2f}")
print(f"Faturamento Total 2023: R$ {total_2023:,.2f}")
print(f"Crescimento em Valor:   R$ {crescimento_total_valor:,.2f}")
print(f"Crescimento Percentual: {crescimento_total_percentual:+.2f}%")
print("-" * 50)
```

### Exercício 6 - Desafio
- No final da reunião de apresentação dos números, seu chefe perguntou:
- E se nos meses de 2023 que a gente vendeu menos do que 2022 a gente tivesse pelo menos empatado com 2022 (ou seja, se nos meses de 2023 em que as vendas foram menores do que o mesmo mês em 2022, o valor de vendas tivesse sido igual a 2022)
- Qual teria sido o nosso crescimento de 2023 frente a 2022?

```python
# PLANEJAMENTO DO EXERCÍCIO 6 (DESAFIO):
# 1. Utilizar os mesmos dicionários do Exercício 5.
# 2. Criar um dicionário novo chamado 'vendas_23_simulado = {}' para rodar a projeção do chefe.
# 3. Percorrer os dados usando o loop 'for mes in vendas_22:'.
# 4. Adicionar uma estrutura de decisão 'if' para comparar os anos:
#    - Se as vendas de 2023 forem MENORES que as de 2022, o mês simulado recebe o valor de 2022 (empate).
#    - Caso contrário, o mês simulado recebe o valor real de 2023.
# 5. Somar o total do dicionário simulado usando sum().
# 6. Recalcular o crescimento percentual do cenário frente ao ano de 2022 original e exibir o resultado.

vendas_22 = {"jan": 15000, "fev": 15500, "mar": 14000, "abr": 16600, "mai": 16300, "jun": 17000}
vendas_23 = {"jan": 17000, "fev": 15000, "mar": 17500, "abr": 16900, "mai": 16000, "jun": 18500}
vendas_23_simulado = {}

# Aplicando a regra do cenário de simulação solicitado
for mes in vendas_22:
    if vendas_23[mes] < vendas_22[mes]:
        # Se 2023 vendeu menos, força o valor a empatar com 2022
        vendas_23_simulado[mes] = vendas_22[mes]
    else:
        # Se vendeu igual ou mais, mantém o recorde real de 2023
        vendas_23_simulado[mes] = vendas_23[mes]

# Calculando os novos totais para a comparação do cenário
total_2022 = sum(vendas_22.values())
total_2023_simulado = sum(vendas_23_simulado.values())

crescimento_simulado_valor = total_2023_simulado - total_2022
crescimento_simulado_percentual = (total_2023_simulado / total_2022 - 1) * 100

print(f"{' CENÁRIO PROJETADO: ELIMINAÇÃO DE QUEDAS ':=^50}\n")
print(f"Dicionário Simulado de 2023: {vendas_23_simulado}\n")
print(f"Faturamento Total 2022 Original:  R$ {total_2022:,.2f}")
print(f"Faturamento Total 2023 Simulado:  R$ {total_2023_simulado:,.2f}")
print(f"Novo Crescimento em Valor:        R$ {crescimento_simulado_valor:,.2f}")
print(f"Novo Crescimento Percentual:      {crescimento_simulado_percentual:+.2f}%")
print(f"\nNota: No cenário real o crescimento foi de +4.39%. No simulado seria de {crescimento_simulado_percentual:+.2f}%.")
print("-" * 50)
```
