## 1. Faturamento do Melhor e do Pior Mês do Ano

Qual foi o valor de vendas do melhor mês do Ano?
E valor do pior mês do ano?

```python
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]


print(meses[(vendas_1sem + vendas_2sem).index(max(vendas_1sem + vendas_2sem))])
print(meses[(vendas_1sem + vendas_2sem).index(min(vendas_1sem + vendas_2sem))])
```

## 2. Continuação

Agora relacione as duas listas para printar 'O melhor mês do ano foi {} com {} vendas' e o mesmo para o pior mês do ano.

Calcule também o faturamento total do Ano e quanto que o melhor mês representou do faturamento total.

Obs: Para o faturamento total, pode usar a função sum(lista) que soma todos os itens de uma lista

```python
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]

vendas_ano=[]
vendas_ano.extend(vendas_1sem + vendas_2sem)

melhor_mes_do_ano = meses[(vendas_ano).index(max(vendas_ano))]
pior_mes_do_ano = meses[(vendas_ano).index(min(vendas_ano))]

melhor_venda_do_ano = max((vendas_ano))
pior_venda_do_ano = min((vendas_ano))

total_vendas_ano = sum(vendas_ano)

print(f"O melhor mês do ano foi {melhor_mes_do_ano} com {melhor_venda_do_ano}")
print(f"O pior mês do ano foi {pior_mes_do_ano} com {pior_venda_do_ano}")
print(f"O faturamento total do ano foi de {total_vendas_ano} e o {melhor_mes_do_ano} representa {(melhor_venda_do_ano/total_vendas_ano):.2%}")
```
